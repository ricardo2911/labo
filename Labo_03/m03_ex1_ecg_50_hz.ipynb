
<br>
    The objective of this exercise is to study the influence of the<br>
    parameterization of the Welch spectral estimator in order to highlight a 50<br>
    Hz perturbation in an ECG signal.<br>


```python
import numpy as np
import pylab as py
py.ion()
py.close('all')
import scipy.signal as sp
```

```python
x = np.genfromtxt('ecg.dat')
fs = 500
```


<br>
    Objective: Compare spectral estimation for different window lengths using<br>
    welch estimation.<br>
    Plot the log spectrum of the signal using windows of 100, 500, 2000.<br>
    

Q: Comment the results.<br>

The first sample, with the shorter sample length, is the smoothest. Some oscillation can be seen on it, outside the window length, which is probably due to truncation of the signal. The more the sample length is increased, the more the graph becomes noisy. A peak is present at 50Hz on the three graphs. In the third graph, the negative amplitude is increased. 

Q: Which windows length is the most suitable for the observation of 50 Hz?<br>

The most suitable window length is the first one (100).



 Q: Why?<br>
 
The first one has the best window length for 50Hz peak observation because it is the one where it is the easiest to distinguish.

```python
f,X_100 = sp.welch(x, nperseg=100, nfft=4096, fs=fs)
f,X_500 = sp.welch(x, nperseg=500, nfft=4096, fs=fs)
f,X_2000 = sp.welch(x, nperseg=2000, nfft=4096, fs=fs)
```

```python
py.figure(1, figsize=[5,8])
py.clf()
py.subplot(3,1,1)
py.plot(f, 10*np.log10(X_100), 'k')
py.grid()
py.xlabel('frequency (Hz)')
py.ylabel('power (dB)')
py.title('PSD with lengh = 100')
```

```python
py.subplot(3,1,2)
py.plot(f, 10*np.log10(X_500), 'k')
py.grid()
py.xlabel('frequency (Hz)')
py.ylabel('power (dB)')
py.title('PSD with lengh = 500')
```

```python
py.subplot(3,1,3)
py.plot(f, 10*np.log10(X_2000), 'k')
py.grid()
py.xlabel('frequency (Hz)')
py.ylabel('power (dB)')
py.title('PSD with lengh = 2000')
```
